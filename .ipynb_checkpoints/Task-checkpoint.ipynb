```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
import xgboost
import matplotlib.pyplot as plt

def accuracy(real_rates, predicted_rates):
    return np.average(abs(real_rates / predicted_rates - 1.0)) * 100.0
```

```python
def preprocessing(df):
    le = LabelEncoder()
    
    le.fit(df.destination_kma)
    df.destination_kma = le.transform(df.destination_kma)
    
    le.fit(df.origin_kma)
    df.origin_kma      = le.transform(df.origin_kma)

    le.fit(df.transport_type)
    df.transport_type  = le.transform(df.transport_type)

    df.fillna(0,inplace=True)
    return df
```

```python
df_train = pd.read_csv('dataset/train.csv')
df_train = preprocessing(df_train)

df_train.head()
```

## GPU Enabled XGBoost Regressor for Feature Engineering

For more information [click here](https://xgboost.readthedocs.io/en/latest/gpu/)

```python
from collections import OrderedDict
#XGBOOST
df_train = pd.read_csv('dataset/train.csv')
le = LabelEncoder()



le.fit(df_train.destination_kma)
df_train.destination_kma = le.transform(df_train.destination_kma)
le.fit(df_train.origin_kma)
df_train.origin_kma      = le.transform(df_train.origin_kma)

le.fit(df_train.transport_type)
df_train.transport_type  = le.transform(df_train.transport_type)

df_train.fillna(0,inplace=True)

model = xgboost.XGBRegressor(colsample_bytree=0.4,
                             tree_method='gpu_hist', 
                             gpu_id=0,
                             gamma=0,                 
                             learning_rate=0.07,
                             max_depth=3,
                             min_child_weight=1.5,
                             n_estimators=10000,                                                                    
                             reg_alpha=0.75,
                             reg_lambda=0.45,
                             subsample=0.6,
                             seed=42) 
every_column_except_y= [col for col in df_train.columns if col not in ['rate','Id','pickup_date']]

model.fit(df_train[every_column_except_y],df_train['rate'])

print(OrderedDict(sorted(model.get_booster().get_fscore().items(), key=lambda t: t[1], reverse=True)))

most_relevant_features= sorted(list( dict((k, v) for k, v in model.get_booster().get_fscore().items() if v >= 10000).keys()),reverse=True)
print(most_relevant_features)
```

## Linear Regression 

```python
from sklearn.linear_model import LinearRegression

X =df_train[most_relevant_features]
y =df_train[["rate"]]
lr = LinearRegression()

lr.fit(X,y)
predicted = lr.predict(X)
mare=accuracy(df_train.rate,predicted[0])
mare = round(mare,2)
print(f'Accuracy of prediction using MARE is {mare}%')

df_train_lr = df_train.copy()
df_train_lr["predicted_rate"]=predicted
df_train_lr.head()
```

```python
df_valid = pd.read_csv('dataset/validation.csv')
df_valid = preprocessing(df_valid)
X =df_valid[most_relevant_features]
predicted = lr.predict(X)
mare=accuracy(df_valid.rate,predicted[0])
mare = round(mare,2)
print(f'Accuracy of prediction is {mare}%')
```

## XGBRegression with bglinear boosting technique


```python
train = df_train[most_relevant_features]

model = xgboost.XGBRegressor(
                             booster="gblinear",    
                             learning_rate=0.05,
                             n_estimators=1000,                                                                    
                             reg_alpha=0.75,
                             reg_lambda=0.45,
                             seed=42
                            )

model.fit(train,df_train.rate,verbose=False)
predicted = model.predict(train)
mare=np.round(accuracy(df_train.rate,predicted),2)
print(f'Accuracy of prediction is {mare}%')

df_train_xgbr = df_train.copy()
df_train_xgbr["predicted_rate"]=predicted
df_train_xgbr
```

```python
df_valid = pd.read_csv('dataset/validation.csv')


df_valid = preprocessing(df_valid)

predicted = model.predict(df_valid[most_relevant_features])
mare=np.round(accuracy(df_valid.rate,predicted),2)
print(f'Accuracy of prediction is {mare}%')
```

## Testing with 2 models

```python
df_test = pd.read_csv('dataset/test.csv')
df_test = preprocessing(df_test)
predicted = model.predict(df_test[most_relevant_features])
df_test_xgbr= df_test.copy()
df_test_xgbr["predicted_rates"]=predicted
df_test_xgbr
```

```python
df_test = pd.read_csv('dataset/test.csv')
df_test = preprocessing(df_test)
predicted = lr.predict(df_test[most_relevant_features])
df_test_lr= df_test.copy()
df_test_lr["predicted_rates"]=predicted
df_test_lr
```
